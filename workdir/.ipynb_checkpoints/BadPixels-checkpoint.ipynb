This document investigates the darks from FUN-040-01-1. Some preliminary conclusions are:

- This dataset is better for readout noise analysis since NDIT=1. From 2-second readouts, I estimate the read noise to be around 4.33 ADU (slightly differing for readout schemes).
- Note that noise drops as the exposure time increases in UTR. This was confusing to me at first. After investigating, I found that the exposure time in UTR mode is controlled by NDSAMPLES. This is fine although the actual read noise is underestimated when NDSAMPLES > 2 and the noise doesn't exactly scale with 1/sqrt(NDSAMPLES).
- Another thing that I am confused is the net exposure time in UTR mode. It should be multiples of frame time (1.59 secs) since NDSAMPLES of 2, 7, 38 are assigned to exposure times of 2, 10, 60 seconds. So, the net exposure times are actually 1.6, 9.6, 58.8 seconds respectively (right?). But, in the header, these values are only noted under MINDIT which is actually minimum integration time for a given NDSAMPLES - not the net exposure time.

```python
RNs = []
RN_err = []
def doAnalysis(select, suffix):
    
    print select, suffix
    NIX_GL.filter(select)
    res, diffs = NIX_GL.getPixelVariance(shift=True)
    res = sqrt(res)
    
    res1d = ravel(res)

    f, ax = subplots()
    hist, bins = histogram(res1d, 200, range=[0, 20])
    bins = (bins[1:] + bins[:-1])/2.

    model = GaussianModel()
    pars = model.guess(hist, x=bins)
    out = model.fit(hist, pars, x=bins)
    print out.fit_report()

    RNs.append(out.params['center'].value)
    RN_err.append(out.params['center'].stderr)
    
    ax.plot(bins, hist, 'k', drawstyle='steps-mid')
    ax.plot(bins, out.best_fit, 'r--')
    ax.set_xlim([0, 10])
    
    ax2 = ax.twinx()
    
    cums = cumsum(hist)/(2048.*2048)*100
    
    ax2.plot(bins, cums, 'g-', drawstyle='steps-mid')
    ax2.tick_params(axis='y', labelcolor='g')
    ax2.set_ylabel('Cumulative (%)', color='g')
    ax2.set_ylim([0, 100])

    limp = [97, 98, 99]
    lims = [min(bins[where(cums > 97)[0]]), min(bins[where(cums > 98)[0]]), min(bins[where(cums > 99)[0]])]

    figure()
    plot(diffs[1:], 'k.')
    title('Stability of the darks (Mean: %4.1f, Std: %4.1f)' % (mean(diffs[1:]), std(mean(diffs[1:]))))
    ylabel('median(Im_i - Im_0)')
    xlabel('Image Number')
    
    f, axs = subplots(1, 3)
    for i in range(len(lims)):
        ax2.plot([lims[i], lims[i]], [0, 100], 'g--')
        
        out = zeros([2048, 2048])
        out[where(res > lims[i])] = 1
        axs[i].title('Outliers %d%%' % limp[i])
        axs[i].imshow(out, cmap=cm.gray, origin='lower')

    show()
    fits.PrimaryHDU(res).writeto('out/Dark_Variance_%s.fits' % suffix, overwrite=True)
    print h.heap()
```

```python
selects = [[12, 62], [62, 112], [112, 162],
          [162, 212], [212, 262], [262, 312],
          [312, 362], [362, 412], [412, 462]]

suffixes = ['UTR_2sec', 'UTR_10sec', 'UTR_60sec',\
            'CDS_2sec', 'CDS_10sec', 'CDS_60sec', \
            'FWLR_2sec', 'FWLR_10sec', 'FWLR_60sec']
```

```python
for i in range(9):
    doAnalysis(slice(selects[i][0], selects[i][1]), suffixes[i])
```

```python
RNs = array(RNs).astype(float)
print RN_err
RN_err = array(RN_err).astype(float)
print RN_err
exps = array([2., 10., 60.])
for i in range(3):
    print "==> Results for %ds darks:" % exps[i]
    print "UTR    - %5.2f +/- %.2f ADUs" % (RNs[i], RN_err[i])
    print "CDS    - %5.2f +/- %.2f ADUs" % (RNs[3+i], RN_err[3+1])
    print "FOWLER - %5.2f +/- %.2f ADUs" % (RNs[6+i], RN_err[6+1])
```
